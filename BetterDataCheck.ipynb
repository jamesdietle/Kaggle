## Functions

```python
def join_df(left, right, left_on, right_on=None):
    if right_on is None: right_on = left_on
    return left.merge(right, how='left', left_on=left_on, right_on=right_on, 
                      suffixes=("", "_y"))
```

```python
def normal(name):
    #2PM_zh.wikipedia.org_all-access_spider
    out= name
    hold=name
    hold= hold.rsplit('_', 2)
    firstsplit= hold[0].rsplit('.', 2)
    secondsplit= firstsplit[0].rsplit('_', 1)
    #print(secondsplit[0])
    #print(secondsplit[1])
    #print((firstsplit[1]))
    #print(hold[len(hold)-2])
    #print(hold[len(hold)-1])
    out = [secondsplit[0],secondsplit[1],firstsplit[1],hold[len(hold)-2],hold[len(hold)-1]]
    return out
```

```python
def wikimedia(name):
    out= name
    hold=name
    hold= hold.rsplit('_', 2)
    pants= hold[0].rsplit('.wiki', 1)
    #print(pants[0])
    #print("wikimedia.org")
    #print("wikimedia.org")
    #print(hold[len(hold)-2])
    #print(hold[len(hold)-1])
    out = [pants[0],"wikimed","wikimedia.org",hold[len(hold)-2],hold[len(hold)-1]]
    return out
```
